```python
no_pages = 55
```

```python
import numpy as np
dataset = pd.DataFrame(columns = ["Restaurants","Rating","Cuisine","Price"])
for each_page in range(1,no_pages):
    
    swiggy_cities="https://www.swiggy.com/city/hyderabad/top-rated-collection?page="
    swiggy_cities= swiggy_cities+str(each_page)
    res = requests.get(swiggy_cities)
    data = res.text
    bs_data = BeautifulSoup(data)
    #soup = BeautifulSoup(data,'html.parser')
    #top_res = soup.find('div',class_='_129-b')
    item_list = bs_data.find_all("div", attrs = {"class":"_1HEuF"})
    for each_item in item_list:
    
        data_dict = {}
        
        title = each_item.find("div" , attrs = {"class":"nA6kb"})
        if(title):
            data_dict["Restaurants"] = title.text
        #else:
            #data_dict["Restaurants"] = np.NaN
        
        rating = each_item.find("div" , attrs = {"class":"_9uwBC wY0my"})
        if(rating):
            data_dict["Rating"] = rating.text
        else:
            data_dict["Rating"] = 0
        
        cuisine = each_item.find("div" , attrs = {"class":"_1gURR"})
        if(cuisine):
            data_dict["Cuisine"] = cuisine.text
        #else:
            #data_dict["Cuisine"] = np.NaN
            
        price=  each_item.find("div" , attrs = {"class":"nVWSi"})
        if(price):
            data_dict["Price"] = price.text
        #else:
            #data_dict["Price"] = 0
        
        delivery_time=  each_item.find("div" , attrs = {"class":"_3Mn31"})
        if(delivery_time):
            data_dict["Delivery Time"] = delivery_time.text[4:11]
        #else:
            #data_dict["Delivery Time"] = 0
        
        dataset = dataset.append(data_dict, ignore_index = True)
        
    
        
```

```python
def parse_amount(amount):
    if(pd.notna(amount)):
        amount = re.sub(r"[^0-9.]", "", str(amount))
        return str(amount)
    else:
        return amount
```

```python
dataset["Price"].apply(parse_amount)
```

```python
dataset.dtypes
```

```python
re_For_one_or_two = "FOR\sTWO"
```

```python
dataset["For_one_or_two"] = dataset["Price"].apply(lambda x : re.search(re_For_one_or_two,x).group() if re.search(re_For_one_or_two,x) != None else np.NaN)
```

```python
dataset
```

```python
dataset[["Price","Rating","Delivery Time"]] = dataset[["Price","Rating","Delivery Time"]].applymap(parse_amount)
```
